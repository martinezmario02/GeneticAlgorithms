```python
# Función para graficar la comparación de costo y tiempo de ejecución
def graficar_comparacion_costo_tiempo(resultados):
    # Extraer los nombres de los algoritmos
    algoritmos = list(resultados.keys())

    # Extraer los costos finales y tiempos de ejecución
    costos = [resultados[alg]['mejor_costo'] for alg in algoritmos]
    tiempos = [resultados[alg]['tiempo_ejecucion'] for alg in algoritmos]

    # Gráfico de Comparación de Costos Finales
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.bar(algoritmos, costos, color=['blue', 'green', 'red'])
    plt.title('Comparación de Costos Finales')
    plt.xlabel('Algoritmo')
    plt.ylabel('Costo')
    plt.grid(True)

    # Gráfico de Comparación de Tiempos de Ejecución
    plt.subplot(1, 2, 2)
    plt.bar(algoritmos, tiempos, color=['blue', 'green', 'red'])
    plt.title('Comparación de Tiempos de Ejecución')
    plt.xlabel('Algoritmo')
    plt.ylabel('Tiempo de Ejecución (segundos)')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Llamar a la función para graficar la comparación de costo y tiempo
graficar_comparacion_costo_tiempo(results)
```
